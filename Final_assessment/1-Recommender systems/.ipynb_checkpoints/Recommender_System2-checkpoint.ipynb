```python
import numpy as np 
import pandas as pd
from surprise import Reader,Dataset,NMF
from surprise.model_selection import cross_validate
from surprise.model_selection import GridSearchCV
```

---
Load rating data

---

```python
rating=pd.read_csv('ratings_small.csv')
rating.head()
```

---
you will need to define a Reader object for Surprise to be able to parse the dataframe.

rating_scale from 0 to 5 because we have minimum rating 0.5 and maximum rating 5

---

```python
reader=Reader(rating_scale=(0, 5))
data=Dataset.load_from_df(rating[['userId','movieId','rating']],reader)
```

---
 the user-item interaction matrix is very sparse, to use it more efficiently we can reduce or compress the user-items interaction matrix into two matrices using a model.  The huge sparse item matrix is decomposed into two smaller dense matrices- a user-factor matrix that has user representations and an item-factor matrix that has item representation using Negative Matrix Factorization(NMF) Techniques for Recommender Systems.
 
 ScreenShot_20211021111423.png

---

we will use grid search on NMF to choose best n_factors (K)

---

```python
param_grid ={'n_factors':list(range(5, 21))}
grid=GridSearchCV(NMF, param_grid, cv=3,refit=True)
grid.fit(data)
```

```python
grid.best_params
```

---
Lets now choose a user of our choice, and view all the films this user has rated.

---

```python
rating[rating['userId']==1]
```

---
and choose a film of our choice, and view all the rating for it from all the users.

---

```python
rating[rating['movieId']==6268]
```

---
The next step is to predict the rating this user would give to a movie that he/she has not rated before.

---

```python
userID,movieID=1,6268
rat=grid.predict(userID,movieID)
print('The system predicts that the user(id={}) will rate the movie(id={}) with {}'.format(userID,movieID,rat[3]))
```
