```python
trans=OrdinalEncoder()
trans.fit(df[['Married/Single']])
df['Married/Single'] = trans.transform(df[['Married/Single']])
trans.fit(df[['House_Ownership']])
df['House_Ownership'] = trans.transform(df[['House_Ownership']])
trans.fit(df[['Car_Ownership']])
df['Car_Ownership'] = trans.transform(df[['Car_Ownership']])
trans.fit(df[['Profession']])
df['Profession'] = trans.transform(df[['Profession']])
trans.fit(df[['CITY']])
df['CITY'] = trans.transform(df[['CITY']])
trans.fit(df[['STATE']])
df['STATE'] = trans.transform(df[['STATE']])
scaler=MinMaxScaler()
df['Income']=scaler.fit_transform(df[['Income']])
```
